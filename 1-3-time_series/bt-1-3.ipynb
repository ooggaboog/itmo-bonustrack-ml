```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://dc-edu.itmo.ru/assets/courseware/v1/3485634dc5fd4834c786b5a3cd5a9a03/asset-v1:ITMO+bonus_track_methods+2023+type@asset+block/task5_447502.csv')
df.index += 1
df
```

```python
df['exp'] = df.ewm(alpha = 0.2, adjust = False).mean()
df
```

```python
print(round(df['exp'][43], 2), round(df['exp'][100], 2))
```

```python
poly = np.polyfit(df.index, df['y'], 1)
slope = poly[0]
intercept = poly[1]
print('Уравнение линейного тренда: y = ', round(slope, 2), 'x + ', round(intercept, 2), sep='')
```

```python
df['lin_trend'] = slope * df.index + intercept
df
```

```python
y = df['y']
y_ = df['lin_trend']

r2_score = 1 - ((y - y_) ** 2).sum() / ((y - y.mean()) ** 2).sum()
print('Коэффициент детерминации =', round(r2_score, 2))
```
