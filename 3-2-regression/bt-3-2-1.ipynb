```python
import pandas as pd
from io import StringIO
import numpy as np
```

```python
data = '''
id,X,Y
1,7,22
2,18,53
3,22,48
4,5,12
5,6,18
6,4,8
7,19,43
8,14,31
9,8,16
10,11,36
'''

df = pd.read_csv(StringIO(data))
df
```

```python
x_mean = df['X'].mean()
print('x_mean =', x_mean)
```

```python
Y_mean = df['Y'].mean()
print('Y_mean =', Y_mean)
```

```python
# коэффициенты линейной регрессии
x = df['X']
Y = df['Y']
n = len(df)

theta1 = np.sum((x - x_mean)*(Y - Y_mean)) / np.sum((x - x_mean)**2)
theta0 = Y_mean - theta1 * x_mean

print('theta1 =', round(theta1, 2))
print('theta0 =', round(theta0, 2))
```

```python
# R^2 - статистика
R2 = 1 - np.sum((Y - theta0 - theta1 * x)**2) / np.sum((Y - Y_mean)**2)
print('R2 =', round(R2, 2))
```
