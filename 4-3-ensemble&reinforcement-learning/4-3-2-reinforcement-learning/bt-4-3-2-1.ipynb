```python
from sympy import symbols, Eq, solve

V1, V2, V3, V4 = symbols('V1 V2 V3 V4')
g = 0.8

# уравнения Белмана
eq1 = Eq(V1, 0.5 * (6 + g * V3) + 0.5 * (0.1 * (2 + g * V2) + 0.7 * (-1 + g * V4) + 0.2 * (3 + g * V3)))
eq2 = Eq(V2, 3 + g * V1)
eq3 = Eq(V3, 0.3 * (-3 + g * V1) + 0.7 * (0.8 * (6 + g * V4) + 0.2 * (1 + g * V3)))
eq4 = Eq(V4, 0.6 * (5 + g * V1) + 0.4 * (-3 + g * V2))

solutions = solve((eq1, eq2, eq3, eq4), (V1, V2, V3, V4))

round_solutions = {}
for key, value in solutions.items():
  round_solutions[key] = round(value, 3)

print(round_solutions)
```
