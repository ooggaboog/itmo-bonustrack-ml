```python
import pandas as pd
import numpy as np
from io import StringIO
```

```python
data = '''
id,X,Y,Class
1,22,86,0
2,78,32,1
3,54,50,1
4,68,80,0
5,18,12,0
6,85,48,0
7,78,10,0
8,50,74,0
9,30,95,0
10,91,12,0
'''

df = pd.read_csv(StringIO(data))
df
```

```python
x = 30
y = 30

df['Evklid'] = np.sqrt((x - df['X'])**2 + (y - df['Y'])**2)
df['Manhat'] = abs(x - df['X']) + abs(y - df['Y'])
df
```

Евклидово расстояние

```python
df = df.sort_values(by='Evklid')
print('Расстояние до ближайшего соседа:', np.round(df['Evklid'].head(1).values, 3)[0])
print('Три ближайшие точки:', df['id'].head(3).values)
print('Класс нового объекта:', df['Class'].head(3).mode()[0])
```

Манхеттенское расстояние

```python
df = df.sort_values(by='Manhat')
print('Расстояние до ближайшего соседа:', np.round(df['Manhat'].head(1).values, 3)[0])
print('Три ближайшие точки:', df['id'].head(3).values)
print('Класс нового объекта:', df['Class'].head(3).mode()[0])
```
