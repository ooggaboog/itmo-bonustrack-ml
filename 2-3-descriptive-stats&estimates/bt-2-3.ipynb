```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Считываем и подготавливаем данные

```python
df = pd.read_csv('salary.csv', delimiter=';')
df
```

Удалим регионы: **Вологодская область, Мурманская обл.**

```python
df = df[(df['REGION_NAME'] != 'Вологодская область') & (df['REGION_NAME'] != 'Мурманская обл.')]
df
```

## Строим вариационный ряд

Отсортируем значения `SALARY` по возрастанию



```python
df = df.sort_values(by='SALARY').reset_index(drop=True)
df.index += 1
df
```

```python
print('X6: ', df['SALARY'][6])
print('X13: ', df['SALARY'][13])
print('X33: ', df['SALARY'][33])
```

## Строим гистограмму распределения

```python
df['SALARY'].hist(bins=10, grid=False)
plt.show()
```

## Описательная статистика

```python
print('Выборочное среднее: ', round(df['SALARY'].mean(), 2))
```

```python
print('Смещенная выборочная дисперсия: ', round(df['SALARY'].var(ddof=0), 2))
```

```python
print('Несмещенная выборочная дисперсия: ', round(df['SALARY'].var(), 2))
```

```python
print('Выборочная медиана: ', round(df['SALARY'].median(), 2))
```
