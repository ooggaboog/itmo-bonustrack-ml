```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('SPFB.RTS-12.18_180901_181231.csv')
df
```

```python
df['datetime'] = df['<DATE>'] + ' ' + df['<TIME>']
df.head()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'], format='%d/%m/%y %H:%M')
df['datetime'].head()
```

```python
df_filtered = df[df['datetime'].dt.strftime('%Y-%m-%d') == '2018-09-28']
df_filtered = df_filtered.set_index('datetime')
df_filtered = df_filtered[['<OPEN>', '<HIGH>', '<LOW>', '<CLOSE>']]
df_filtered.columns =['open', 'high', 'low', 'close']
df_filtered.head()
```

```python
to_plot = df_filtered.resample('1H').agg({'open':'first',
                                     'high':'max',
                                     'low':'min',
                                     'close':'last'})
to_plot
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=to_plot.index,
                open=to_plot['open'],
                high=to_plot['high'],
                low=to_plot['low'],
                close=to_plot['close'])])

fig.show()
```
