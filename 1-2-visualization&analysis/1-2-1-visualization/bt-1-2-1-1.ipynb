```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('var123719.xlsx')
df
```

```python
to_plot = pd.DataFrame(df['Choice_1'].value_counts())
to_plot = to_plot.sort_index()
to_plot
```

```python
plt.bar(x=to_plot.index, height=to_plot['count'])
plt.xticks(rotation=45)
plt.show()
```

```python
flatten_to_plot = pd.DataFrame(pd.DataFrame(df.values.reshape(-1))[0].value_counts())
flatten_to_plot = flatten_to_plot.sort_index()
flatten_to_plot
```

```python
plt.pie(x=flatten_to_plot['count'], labels=flatten_to_plot.index, autopct='%1.1f%%')
plt.show()
```
