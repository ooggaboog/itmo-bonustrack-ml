```python
import pandas as pd

df = pd.read_csv('https://dc-edu.itmo.ru/assets/courseware/v1/cb35bafc7d16bcf4dce066850ccecb12/asset-v1:ITMO+bonus_track_methods+2023+type@asset+block/task4_1_625176.csv',
                       delimiter=',',
                       decimal='.', index_col = 'ID')
df
```

```python
df = df[df['INTERNET'] != 0]
```

```python
import numpy as np

df = 1 - np.exp(1 - df / df.min())
df
```

```python
df['func'] = df['DISTANCE'] + df['STOP_COUNT'] + df['COST'] + df['FITNESS'] * 0.5 - df['INTERNET'] * 0.2
df.sort_values(by=['func']).head(3)
```
